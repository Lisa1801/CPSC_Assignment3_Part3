```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
mean_price = df['SalePrice'].mean()
std_price = df['SalePrice'].std()

print("Mean Sale Price: ", mean_price)
print("Standard Deviation of Sale Price: ", std_price)
```

```python
df.isnull().sum()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.show()
```

```python
#Graph for the main variable "Sale Price"

plt.hist(df['SalePrice'])
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Hypothesis1
#Sale price versus Lotarea
sale_price = df.groupby('LotArea')['SalePrice'].mean()
plt.scatter(sale_price.index, sale_price.values)
plt.xlabel('Lot Size')
plt.ylabel('Sale Price')
plt.show()
```

```python
#From above graph it is clear that the variables have a postive relation. And mostly houses are sold upto the lot size 50000.
```

```python
#Hypothesis2
#Sale price versus neighborhood



sale_price= df.groupby('Neighborhood')['SalePrice'].mean()
plt.scatter(sale_price.index, sale_price.values)
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price')
plt.show()
```

By this graph it is clearly seen that there is no correlation between both the variables and the selling of houses does not 
depend on the neighborhood.

```python
#Hypotheis 3
#Sale price versus Overall Quality

sale_price= df.groupby('OverallQual')['SalePrice'].mean()
plt.bar(sale_price.index, sale_price.values)
plt.xlabel('OverallQual')
plt.ylabel('SalePrice')
plt.show()
```

It clearly shows the relation is positive and the sale price depends on the overall quality of houses.
